## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Creating a DataFrame of no. of kills in each Country.
pd.DataFrame(df.groupby('country_txt')['nkill'].count()).reset_index()
````

````python
# Getting a value of no. of kills by terrorist organisations.
df.groupby('gname')['nkill'].nunique().sort_values(ascending=False)[:10]
````

````python
# Creating DataFrame of no. of attack by every terrorist organisations in each country.
pd.DataFrame(df.groupby('country_txt')['gname'].value_counts().reset_index(name="Count"))
````

````python
# Top 10 Terriorist Organisations attack in Pakistan
country_df = df.explode('country_txt')
pd.DataFrame(country_df[country_df['country_txt'].isin(['Pakistan'])].explode('nkill')['gname'].value_counts().reset_index(name="Attack Information"))[:10]
````

````python
# Different types of successful attacks
pd.DataFrame(df.groupby('attacktype1_txt')['success'].count()).reset_index()
````

````python
# Finding out different types of attacks in countries.
pd.DataFrame(df.groupby('country_txt')['attacktype1_txt'].value_counts().reset_index(name="Count"))
````

````python
# No. of people killed and wounded every year from 1970 to 2017 by terrorist organisations.
df.groupby('iyear').agg({"gname": "sum", "nkill": "count", "nwound": "count"})
````

````python
# Getting Unique Countries
print(df['country_txt'].unique())
print(" ")
# Getting Unique Countries Count
print("Unique Country Count is ",df['country_txt'].nunique(),".")
````
